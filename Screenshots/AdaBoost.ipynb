# AdaBoost

### <u>Description:</u>
AdaBoost (short for Adaptive Boosting) is a machine learning algorithm used to improve the performance of weak models by combining them into a strong one.

Primarily uses decision trees such as the image below:

<div style="text-align: center;">
    <img src="image.png" width="600">
</div>

AdaBoost typically uses simple models called decision stumps, which are decision trees with only one split. It is a specific type of boosting algorithm that improves performance by assigning weights to training data. Initially, all data points are given equal weight, but as training progresses, misclassified points are given higher weights while correctly classified points receive lower weights. This forces subsequent models to focus more on difficult cases. In the end, all models are combined using a weighted vote to produce a highly accurate final model.

### <u>Advantages</u>
* High accuracy by combining many weak learners into a strong model
* Works well with simple models like decision stumps
* Reduces bias by focusing on previous mistakes
* Improves performance by adapting to hard-to-classify data points
* Flexible (can be used for classification and regression variants)





### <u>Disadvantages:</u>
* Sensitive to outliers and noisy data
* Can overfit if data contains mislabeled examples
* Trains sequentially, so it is slower than single models
* Performance depends on weak learners being slightly better than random
* Less interpretable because it combines many models
